#### Load Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Load Dataset

```python
df = pd.read_csv(r"C:\Users\Sree\Downloads\MSFT.csv")
df.head()
```

#### Calculate the Moving Average

```python
df['MV'] = df['Close'].rolling(window = 5).mean()
```

#### Plot the Moving average

```python
x = df['Date'][-100:] # last 100 days of moving average
y = df['Close'][-100:]
y1= df['MV'][-100:]
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,label = 'Close Price',color = 'green')
plt.plot(x,y1,'r--',label = 'Moving Average')
plt.legend()
plt.title('Moving Average Of close price')
plt.xticks([])
plt.show()
```
